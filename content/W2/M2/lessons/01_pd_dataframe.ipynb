## Read data in

Data can be loaded into a Pandas DataFrame from various sources and different formats:

1. **Databases**: relational (SQL) and unstructured (NoSQL)
2. **Surveys**: Google Forms, SurveyMonkey
3. **APIs**: Social Media Platforms, Open Street Maps, Flights, ..etc.
4. **Web Scraping**: BeautifulSoup, Scrapy, Selenium
5. **Raw Files**: csv, excel, txt, json, xml, pdf, ..etc.

Pandas provides the [`read_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html#pandas.read_csv "pandas.read_csv") function to read data stored as a csv file into a pandas `DataFrame`. pandas supports many different file formats or data sources out of the box (csv, excel, sql, json, parquet, …), each of them with the prefix `read_*`.

![](../assets/pandas/pandas_io.png)

See: [IO tools (text, CSV, HDF5, …)](https://pandas.pydata.org/docs/user_guide/io.html) for a table containing available readers and writers.

```python
help(pd.read_csv)
```

The result of a `read_*` operation is a `DataFrame` (usually abbreviated as `df`), which is essentially a **table of rows and columns**.

### Select specific rows

![](../assets/pandas/subset_rows.png)

To select rows based on a conditional expression, use a condition inside the selection brackets `[]`, like so:

```python
above_35 = df_titanic[df_titanic["Age"] > 35]
```

```python
above_35.head()
```

Notice the index shows: `1`, `6`, `11`, `13` and `15`, which are the location of these rows in the original `DataFrame`.

#### Mask

The condition inside the selection brackets `titanic["Age"] > 35` checks for which rows the `Age` column has a value larger than 35. It is also known as a **boolean mask**, since it results in a `Series` of `True`/`False` values:

```python
df_titanic["Age"] > 35
```

Only rows for which the value is `True` will be selected.

### combine row selection conditions

When combining multiple conditional statements, be careful:

1. each condition must be surrounded by parentheses `()`
2. you can not use `or`/`and` but need to use the or **operator `|`** and the **and operator `&`**.


```python
# Condition 1
mask_1 = df_titanic["Age"] > 35

# Condition 2
mask_2 = df_titanic["Pclass"].isin([2, 3])

# Combine conditions
mask = mask_1 & mask_2

# Apply the mask to the DataFrame
above_35_df = df_titanic[mask]
above_35_df.head()
```

### select specific rows AND columns

![](../assets/pandas/subset_rows_and_columns.png)

Let's say, we're interested in the names of the passengers older than 65 years.

When selecting both rows AND columns, We can't just use the brackets `[]` notation directly, rather, we need to preceed add a `loc` or `iloc` method:

1. Use `loc` for label-based selection (using row/column names)
2. Use `iloc` for position-based selection (using table positions)

#### `loc`: label-based selection

```python
seniors_names = df_titanic.loc[df_titanic["Age"] > 65, "Name"]
seniors_names.head()
```

When using either `loc` or `iloc`:

1. the part before the comma is the **rows** you want
2. the part after the comma is the **columns** you want to select

Let's demonstrate combining conditions and multiple column selection:

```python
condition = (df_titanic["Age"] > 65) & (df_titanic['Sex'] == 'male')
columns = ["Name", 'Pclass']

seniors = df_titanic.loc[condition, columns]
seniors.head()
```

To demonstrate using the result, let's say we want to know the first names these people have:

```python
# convert Series to a Python list
names = seniors['Name'].to_list()

# show only first names
first_names = [x.split(',')[1].strip() for x in names]
first_names
```

#### `iloc`: position-based selection

```python
df_titanic.iloc[2:5, 0:3]
```

- Row-wise, we are selecting from `2` to `5`
- Column-wise, we are selecting from `0` to `3`

## Summary: Indexing / selection

The basics of indexing are as follows:

| Operation                      | Syntax          | Result    |
| ------------------------------ | --------------- | --------- |
| Select column                  | `df[col]`       | Series    |
| Select row by label            | `df.loc[label]` | Series    |
| Select row by integer location | `df.iloc[loc]`  | Series    |
| Slice rows                     | `df[5:10]`      | DataFrame |
| Select rows by boolean vector  | `df[bool_vec]`  | DataFrame |

Source: https://pandas.pydata.org/docs/user_guide/dsintro.html#indexing-selection

## Create new columns derived from existing columns

![](../assets/pandas/create_new_columns.png)

Let's demonstrate calculate the BMI given weight and height of each athlete:

```python
# Best practice: use pathlib to handle file paths
from pathlib import Path

# Set the path to the directory containing the datasets
data_dir = Path('../../data')

# Debugging
print(data_dir.resolve())
print(data_dir.exists())
```

```python
data_socr = data_dir / 'SOCR-HeightWeight.csv'

# Debugging
print(data_socr.exists())
print(data_socr.resolve())
```

**BMI formula**:

$$
BMI = \frac{\text{Weight}}{\text{Height}^2}
$$

```python
df_socr = pd.read_csv(data_socr)
df_socr.head()
```

### Vector-scalar operations

Since the Height and Weight are in pounds, and the formula assumes **Metric System**, we need to convert the **Imperial System** numbers first:

```python
df_socr['Height(meters)'] = df_socr['Height(Inches)'] * 0.0254
df_socr['Weight(kilograms)'] = df_socr['Weight(Pounds)'] * 0.453592
df_socr.head()
```

Notice how we are multiplying a **scalar** like `0.0254` by a **vector** (`Series`). This is done **element-wise** (on each element), known as **Broadcasting** the operation.

### Row-wise calculations

Then actually do the calculation:

```python
df_socr['BMI'] = df_socr['Weight(kilograms)'] / (df_socr['Height(meters)'] ** 2)
df_socr.head()
```

Notice how this involves calculation **across two columns of the corresponding row**.

### `apply` functions to calculate values

**BMI table for adults**. This is the World Health Organization's (WHO) recommended body weight based on BMI values for adults. It is used for both men and women, age 20 or older:

| Classification    | BMI range - kg/m2 |
| ----------------- | ----------------- |
| Severe Thinness   | < 16              |
| Moderate Thinness | 16 - 17           |
| Mild Thinness     | 17 - 18.5         |
| Normal            | 18.5 - 25         |
| Overweight        | 25 - 30           |
| Obese Class I     | 30 - 35           |
| Obese Class II    | 35 - 40           |
| Obese Class III   | > 40              |

Source: https://www.calculator.net/bmi-calculator.html

We can code this mapping as a pure Python function:

```python
def get_bmi_category(bmi: float) -> str:
    if bmi < 16:
        return 'Severe Thinness'
    elif bmi < 17:
        return 'Moderate Thinness'
    elif bmi < 18.5:
        return 'Mild Thinness'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    elif bmi < 35:
        return 'Obese Class I'
    elif bmi < 40:
        return 'Obese Class II'
    else:
        return 'Obese Class III'
```

Then use `apply` to process each value in the column, as follows:

```python
df_socr['classification'] = df_socr['BMI'].apply(get_bmi_category)
```

Let's sample (randomly) 10 rows to see their classifications:

```python
df_socr.sample(10)
```

Based on that, one could do lots of things like, `value_counts`:

```python
df_socr['classification'].value_counts()
```

### Dropping columns

We might choose to drop the previous columns as well:

```python
df_socr = df_socr.drop(
    columns=['Height(Inches)', 'Weight(Pounds)'],
)
df_socr.head()
```

### Rename columns

Let's say we want to rename from `meters` to `m`, and from `kilograms` to `kg` instead:

```python
df_socr = df_socr.rename(
    columns={
        'Height(meters)': 'Height(m)',
        'Weight(kilograms)': 'Weight(kg)',
    }
)
df_socr.head()
```

## Key Takeaways

- **DataFrames** are core to pandas: they provide labeled, 2-dimensional data structures, allowing flexible and powerful manipulation and analysis of tabular data.

- **Reading Data:** Easily load data from various sources (CSV, Excel, SQL, web, etc.) with functions like `pd.read_csv()`.

- **Quick Inspection:** Use methods such as `.head()`, `.tail()`, `.info()`, and `.dtypes` to quickly get an overview of your data.

- **Selecting Data:**
  - Select columns as `df["col"]` or `df[["col1", "col2"]]`.
  - Select rows using slicing or with boolean conditions, e.g., `df[df['Age'] > 35]`.
  - Use `.loc[]` for selection by labels and `.iloc[]` for selection by integer position.

- **Vectorized Operations:** Arithmetic and logical operations between columns or between columns and scalars are applied elementwise (“broadcasting”).

- **Modifying DataFrames:**
  - Add new columns by assignment (e.g. `df["BMI"] = ...`).
  - Apply functions on values of a column using `apply(func)`
  - Drop columns with `df.drop(columns=[...])`.
  - Rename columns with `df.rename(columns={...})`.

- **Chaining Conditions:** Combine multiple boolean conditions with `&` (and), `|` (or), using parentheses as needed.

These core tools and concepts are essential for effective data exploration, cleaning, and analysis using pandas.
