```python
import matplotlib.pyplot as plt
import numpy as np

# Data generation
x = np.linspace(0.1, 10, 100)
train_error = 1/x + 0.1
val_error = 1/x + 0.05 * (x - 4)**2 + 0.5

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Common settings
titles = ['Under-fitting', 'Optimal Model', 'Over-fitting']
colors = ['blue', 'red']

for i, ax in enumerate(axes):
    ax.plot(x, train_error, label='Train Error', color=colors[0], linewidth=2)
    ax.plot(x, val_error, label='Val Error', color=colors[1], linewidth=2)
    ax.set_title(titles[i], fontweight='bold')
    ax.set_xlabel('Complexity')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.set_ylim(0, 4)

# Subplot 1: Under-fitting
axes[0].annotate('High Bias\nBoth High', xy=(1, 1.5), xytext=(3, 3),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
axes[0].set_ylabel('Error')

# Subplot 2: Optimal
axes[1].annotate('Lowest Point\nBoth Low', xy=(4, 0.75), xytext=(4, 2.5),
                 arrowprops=dict(facecolor='green', shrink=0.05, width=1, headwidth=5), ha='center')

# Subplot 3: Over-fitting
axes[2].fill_between(x, train_error, val_error, where=(x > 5), color='gray', alpha=0.2)
axes[2].annotate('Widening Gap\nHigh Variance', xy=(8, 1.8), xytext=(5, 3),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))

plt.suptitle('Model Performance Breakdown: Ensure data is representative', fontsize=14, y=1.05)
plt.tight_layout()
plt.legend(loc='upper right')

plt.savefig('model_analysis_subplots.png')
```
